```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import geopy
import geopy.distance
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
data = pd.read_csv('ALL_SPB_STOPS.csv', sep=';', encoding='cp1251')
data
```

```python
parnas_coord = (round(float(data[data['STOP_NAME'] == "Парнас"].LATITUDE), 6), (round(float(data[data['STOP_NAME'] == "Парнас"].LONGITUDE), 6)))
parnas_coord
```

```python
parnas = pd.DataFrame({'ID_STOP':[], 'LATITUDE':[], 'LONGITUDE':[]})
parnas
```

```python
for index, row in data.iterrows():
    if row.ID_STOP != int(data[data["STOP_NAME"] == "Парнас"].ID_STOP) :
        area = (row.LATITUDE, row.LONGITUDE)
        if geopy.distance.geodesic(parnas_coord, area).m <= 1000 :
            parnas.loc[len(parnas)] = [row.ID_STOP, row.LATITUDE, row.LONGITUDE]
```

```python
parnas.merge(data[['ID_STOP', 'STOP_NAME']], how='left')
```

```python
link = linkage(parnas[['LATITUDE', 'LONGITUDE']], 'ward', 'euclidean')
```

```python
dn = dendrogram(link)
```

```python
parnas['cluster'] = fcluster(link, 3, criterion='maxclust')
parnas.groupby('cluster').size()
```

```python
parnas.merge(data[['ID_STOP', 'STOP_NAME']], how='left')
```
